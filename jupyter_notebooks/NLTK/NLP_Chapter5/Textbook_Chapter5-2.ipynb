```python
#Find out the tag and corresponding count in the Brown corpus
from nltk.corpus import brown
brown_news_tagged = brown.tagged_words(categories='news', tagset='universal')
tag_freqdist = nltk.FreqDist(tag for (word, tag) in brown_news_tagged)
```

```python
#Find out all the keys in the brown corpus
tag_freqdist.keys()
```

```python
word_tag_pairs = nltk.bigrams(brown_news_tagged)
list(word_tag_pairs)[0:20]
```

```python
word_tag_pairs = nltk.bigrams(brown_news_tagged)
#find out the common tags prior to the 'NOUN' tag
list(nltk.FreqDist(a[1] for (a, b) in word_tag_pairs if b[1] == 'VERB'))
```

```python
#Find the most frequent noun tags
def findtags(tag_prefix, tagged_text):
    #return [n for n in nltk.corpus.brown.tagged_words(categories='news') if n[1] == 'NN']
    cfd = nltk.ConditionalFreqDist((tag, word) for (word, tag) in tagged_text if tag.startswith(tag_prefix))
    return cfd
cd = findtags('NN', nltk.corpus.brown.tagged_words(categories='news'))
cd.conditions()
```
