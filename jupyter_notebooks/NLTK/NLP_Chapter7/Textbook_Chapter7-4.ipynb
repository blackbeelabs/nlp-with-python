# 7.4 Recursion in Linguistic Structure / Trees

```python
grammar = r"""
            NP: {<DT|JJ|NN.*>+}
            PP: {<IN|NP>}
            VP: {<VB.*><NP|PP|CLAUSE>+$}
        CLAUSE: {<NP|VP>}
          """
```

```python
import nltk
cp = nltk.RegexpParser(grammar)
```

```python
sentence = [("Mary", "NN"), ("saw", "VBD"), ("the", "DT"), ("cat", "NN"),
("sit", "VB"), ("on", "IN"), ("the", "DT"), ("mat", "NN")]
```

```python
cp.parse(sentence)
```

```python
sentence2 = [("John", "NNP"), ("thinks", "VBZ"), ("Mary", "NN"), 
            ("saw", "VBD"), ("the", "DT"), ("cat", "NN"), 
            ("sit", "VB"), ("on", "IN"), ("the", "DT"), ("mat", "NN")]
cp.parse(sentence2)
```

```python
#Add 1 loop to specify the no. of times the patterns should be run
cp2 = nltk.RegexpParser(grammar, loop=2)
```

```python
#Trees and tree structure in NLTK
tree1 = nltk.Tree('NP', ['Alice'])
tree1
```

```python
tree2 = nltk.Tree('NP', ["the", "tree"])
tree2
```

```python
tree3 = nltk.Tree('VP', ["chased", tree2])
tree3
```

```python
tree4 = nltk.Tree('S',[tree1, tree3])
tree4
```

```python
tree4[1].label()
```

```python
def traverse(t):
    try:
        t.label()
    except AttributeError:
        print(t),
    else:
        print("t",t.node)
        for child in traverse(child):
                print(")")
t = nltk.Tree("(S (NP Alice) (VP chased (NP the tree)))")
traverse(t)
```
